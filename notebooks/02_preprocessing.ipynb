## 1. Ejemplos de normalización individual

```python
examples = [
    'Click here: https://evil-bank.com/login para verificar',
    'Envía tu info a hacker@darkweb.com URGENTE',
    'Llama al 55-1234-5678 o al +52 (55) 9876 5432',
    'Recibirás $10,000 pesos en tu cuenta bancaria',
    'URGENT!!! ACT NOW!!! You WON!!!',
]

for ex in examples:
    print(f'Original : {ex}')
    print(f'Limpio   : {clean_text(ex)}')
    print()
```

## 2. Verificación de tokens preservados

```python
test_cases = {
    'URL preservada como token': 'Verifica en http://fake-site.com',
    'Email preservado como token': 'Contacta a admin@fraud.org',
    'Teléfono preservado': 'Llama al 5512345678',
    'Dinero preservado': 'Gana $5,000 dólares',
    'Exclamación preservada': 'Actúa AHORA!!!',
    'Texto vacío': '',
    'Solo espacios': '   ',
}

for description, text in test_cases.items():
    result = clean_text(text)
    print(f'{description}:')
    print(f'  Entrada: {repr(text)}')
    print(f'  Salida : {repr(result)}')
    print()
```

## 3. Preprocesamiento de un DataFrame completo

```python
sample_df = pd.DataFrame({
    'message': [
        'ALERT: Your BBVA account is BLOCKED! Click https://bbva-fake.com NOW',
        'Felicidades ganaste $50,000. Envía tu tarjeta a premio@sorteo.net',
        'Hola, ¿cómo estás? Te llamo mañana.',
        'Transfiere $2,500 a la cuenta 012345678901 antes de las 18:00',
        '',  # vacío — debe eliminarse
    ],
    'label': ['fraudulent', 'fraudulent', 'legitimate', 'fraudulent', 'legitimate'],
})

print('Antes del preprocesamiento:')
print(sample_df)
print()

processed_df = preprocess_dataframe(sample_df)
print('Después del preprocesamiento:')
print(processed_df)
```
